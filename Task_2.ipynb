**11.** Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
def sample(x, c):
    assert len(x) > 0
    
    s = np.sum(x)
    res = []
    for _ in range(c):
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val:
            cur += x[idx]
            idx += 1
            
        res.append(idx)
    return res

            
sample([50, 3, 1, 7, 20], 5)    
```

Реализуйте данный алгоритм более оптимальным способом, без использования циклов.

```python
## your code
# используется функция cumsum - прибавляет для каждого следующего элемента массива значение из предыдущего, собственно, происходит то же самое, что и в цикле в вышеописанной функции sample, а дальше значения икс сравниваются со всеми значениями вектора val, затем результаты суммируются, чтобы получить ответ
def sample2(x, c):
    assert len(x) > 0

    s = np.sum(x)
    val = s * np.random.random(c).reshape((c, 1))
    x = np.array(x).cumsum().reshape((1, c))
    res = (x < val)
    return res.sum(axis=1)


sample2([50, 3, 1, 7, 20], 5)
```
